```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
```

```python
df1=pd.read_csv(r"../data/student-mat.csv", sep=";")
df1 = df1.copy()

df2=pd.read_csv(r"../data/student-por.csv", sep=";")
df2 = df2.copy()


```

```python
#get categorical data and split them into dummy variables
categorical_cols1 = df1.select_dtypes(include="object").columns

categorical_cols2 = df2.select_dtypes(include="object").columns


df1_encoded = pd.get_dummies(df1, categorical_cols1)
df2_encoded = pd.get_dummies(df2, categorical_cols2)
```

So now our new data that we would be working on would be the encoded data for both datasets

```python
#i have to merge both dataframes together to avoid training and testing complications

merged_data = pd.merge(df1_encoded, df2_encoded, left_index=True, right_index=True)
list(merged_data.columns)
```

```python
features = merged_data.drop(["G1_x", "G2_x", "G3_x", "G1_y", "G2_y", "G3_y"], axis=1)

label = merged_data[["G1_x", "G2_x", "G3_x", "G1_y", "G2_y", "G3_y"]]
```

```python
#scaling the features

scaler = StandardScaler()
features = scaler.fit_transform(features)


```

```python
#data splitting
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.2, random_state=42)


```

```python
clf = MultiOutputRegressor(LinearRegression())

#initializing the model
```

```python
clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)



print(accuracy)
```
